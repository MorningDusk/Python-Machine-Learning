```python
import pandas as pd
batter_df =  pd.read_csv(batter)
picher_df =  pd.read_csv(picher)
```

```python
import matplotlib.pyplot as plt
```

```python
picher_df.head(3)
```

```python
# 연봉분포를 출력
picher_df['연봉(2018)'].hist(bins=100)
```

```python
# boxplot
picher_df.boxplot(column=['연봉(2018)'])
```

```python
# 회귀분석에 사용할 피처
picher_df_feature = picher_df.loc[:,'승':]
picher_df_feature.head(3)
```

```python
# 각 피처별 데이터 분포 - hist
def plot_hist_cols(df):
  print(len(df.columns))
  plt.figure(figsize=(20,16))
  for idx, col in enumerate( df.columns ):
    plt.subplot(4,5,idx+1)
    plt.hist(df[col],bins = 50)
    plt.title(col)
  plt.show()

plot_hist_cols(picher_df_feature)
```

```python
train = picher_df_feature.drop(columns=['연봉(2018)'])
target = picher_df_feature['연봉(2018)']
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
train_scaled = ss.fit_transform(train)
train_scaled = pd.DataFrame(train_scaled, columns=train.columns)
train_scaled.head()
```

```python
# 범주형 데이터인 팀명을 one-hot-encoding
pd.options.display.max_columns = None
train_scaled_df = pd.concat([ picher_df['팀명'],train_scaled], axis=1)
train_scaled_df = pd.get_dummies(train_scaled_df)
train_scaled_df.head(2)
```

```python
target
```

```python
x = train_scaled_df.values.tolist()
y = target.tolist()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import xgboost as xgb
model_ml = xgb.XGBRegressor()
model_ml.fit(x_train,y_train)
```

```python
# 피처 중요도 출력
model_ml.feature_importances_, train_scaled_df.columns
```

```python
import numpy as np
```

```python
indexs =  np.argsort(model_ml.feature_importances_)
indexs = indexs[::-1]
train_scaled_df.columns[indexs]
```

```python
import statsmodels.api as sm
```

```python
len(x_train[0])
```

```python
X_train = sm.add_constant(x_train)
model = sm.OLS(y_train,X_train).fit()
len(X_train[0]), len(x_train[0])
```

```python
model.summary()
```

```python
# 회귀 계수 마지막에 절편이 추가
plt.figure(figsize=(20,5))
coef = model.params.tolist()
plt.bar(train_scaled_df.columns,coef[:-1] )
plt.xticks(rotation=90)
plt.show()
```

```python
model.params
```

```python
model_ml.score(x_test,  y_test)
```

```python
# rmse
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test, model_ml.predict(x_test)))
```
