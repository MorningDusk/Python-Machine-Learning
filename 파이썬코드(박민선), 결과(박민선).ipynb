```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plotResult(x, y):
    plt.scatter(x[y == 0, 0], x[y == 0, 1], c='blue', marker='o', label='clu-1')
    plt.scatter(x[y == 1, 0], x[y == 1, 1], c='red', marker='s', label='clu-2')
    plt.legend()
    plt.show()
```

```python
x,y = make_moons(n_samples=300, noise=0.05, random_state=0)
plt.scatter(x[:,0], x[:,1], marker='o', c=y, s=50, edgecolors='k', linewidths=1)
plt.xlabel('$x_2$')
plt.ylabel('$y_2$')
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=15)
clusters = dbscan.fit_predict(x_train_scaled)
```

```python
x_test_scaled = ss.transform(x_test)
test_clusters = dbscan.fit_predict(x_test_scaled)
```

```python
plotResult(x_train_scaled, clusters)
```

```python
silhouette_avg = silhouette_score(x_train_scaled, clusters)
```
