```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
import graphviz

mglearn.plots.plot_animal_tree()
```

```python
from sklearn.datasets import make_moons

x,y = make_moons(n_samples=100, noise=0.25, random_state=3)
fig, ax = plt.subplots()
mglearn.discrete_scatter(x[:,0], x[:,1], y, ax=ax)
ax.set_xlabel('Feature 0')
ax.set_ylabel('Feature 1')
ax.legend(['Class 0', 'Class 1'], loc='best')
```

```python
from mglearn.plot_interactive_tree import plot_tree, tree_image

fig, ax = plt.subplots(1, 2, figsize=(12,4), subplot_kw = {'xticks':(), 'yticks':()})
tree = plot_tree(x, y, max_depth=1, ax=ax[0])
ax[1].imshow(tree_image(tree))
ax[1].set_axis_off()
fig.tight_layout()
```
