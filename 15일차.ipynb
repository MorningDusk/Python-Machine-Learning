# 시각화

```python
import matplotlib.pyplot as plt
```

```python
x = range(100)
y = range(100)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
x1 = range(100)
x2 = range(100)
y1 = [np.cos(i) for i in x]
y2 = [np.sin(i) for i in x]

plt.plot(x1, y1, label='cos')
plt.plot(x2, y2, label='sin')
plt.legend()
plt.show()
```

```python
# 축을 설정해서 시각화
x_list = [x1, x2]
y_list = [y1, y2]
tile_list = ['cos', 'sin']
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].plot(x1, y1)
ax[1].plot(x2, y2)
ax[0].set(title='cosine', xlabel='x', ylabel = 'y')
ax[1].set(title='sine', xlabel='x', ylabel='y')
plt.show()
```

```python
plt.figure(figsize=(10,5))
for i in range(2):
    plt.subplot(1,2, i+1)
    plt.plot(x_list[i], y_list[i])
plt.show()
```

```python
# 2 by 2 화면을 분할
x = np.linspace(-1, 1, 100)
```

```python
y = [[np.sin(x), np.cos(x)], [np.tan(x), np.exp(x)]]
```

```python
# 축 ax는 행렬
_, ax = plt.subplots(2,2)
for row in range(2):
    for col in range(2):
        ax[row, col].plot(x,y[row][col])
plt.show()
```

```python
y = [np.sin(x), np.cos(x), np.tan(x), np.exp(x)]
for i in range(2*2):
    plt.subplot(2,2, i+1)
    plt.plot(x, y[i])
plt.show()
```

# 색상

```python
x = range(100)
y = range(100)
plt.plot(x,y, color='r')
plt.plot(x, np.array(y)+10, c='#FFFF00')
plt.show()
```

```python
x = range(100)
y = range(100)
plt.plot(x,y, color='r', linestyle='dashed', label='line1')
plt.plot(x, np.array(y)+10, c='#FFFF00', ls='dotted', label='line2')
plt.title('two lines')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color='r', linestyle='dashed')

plt.subplot(1,2,2)
plt.plot(x, np.array(y)+10, c='#FFFF00', ls='dotted')
plt.show()
```

# 산점도

```python
data1 = np.random.rand(512,2)
data2 = np.random.rand(512,2)
area = np.pi * (np.random.rand(50))**2
plt.scatter(data1[:,0], data2[:,1], c='b', marker='x')
plt.scatter(data2[:,0], data2[:,1], c='r', marker='o')
plt.show()
```

```python
data1 = np.random.rand(512,2)
area = np.pi * (np.random.rand(512)*10)**2
plt.scatter(data1[:,0], data1[:,1], c=np.random.rand(512), marker='o', s=area, alpha=0.5)
plt.show()
```

# 막대그래프 - 범주형 데이터(카테고리)

```python
from random import sample
```

```python
# 3 * 4
data = sample(range(100), 12)
data = np.array(data).reshape(3,4)
data
```

```python
# x의 시작좌표
x = np.arange(0,8,2)
plt.bar(x, data[0], width=0.5)
plt.bar(x+0.5, data[1], width=0.5)
plt.bar(x+1, data[2], width=0.5)
plt.xticks(x+0.5, ('a','b','c','d'))
```

# 누적막대그래프

```python
x = np.arange(data.shape[1])
color_list = ['r', 'g', 'b']
data_label = ['A', 'B', 'C', 'D']
```

```python
plt.bar(x, data[0], bottom=19)
plt.bar(x, data[1])
```

```python
for i in range(3):
    print(f"bottom = {np.sum(data[:i], axis=0)}")
    plt.bar(x, data[i], bottom=np.sum(data[:i],axis=0), color=color_list[i], label=data_label[i])
plt.legend()
plt.show()
```

# boxplot 4분위수를 시각화, 데이터의 분포와 밀집 정도를 표현 또는 분산
    최소값 최대값은 실제 최대 최소가 아니라, 25% Q1, 50% Q2, 75% Q3
    IRQ = Q3 - Q1
    최소값: Q1 - IRQ
    최대값: Q3 + IRQ
    최소 최대 범위를 벗어난 데이터는 outlier(이상치)

```python
data = np.random.randn(100,5)
plt.boxplot(data)
plt.show()
```

# 회귀그래프
    머신러닝의 회귀와 같이 회귀알고리즘을 적용해서 선형(직선)회귀 선을 그래프와 함께 그린다

```python
# 새로운 데이터를 로드해서 확인
tips = sns.load_dataset('tips')
sns.regplot(x='total_bill', y='tip', data=tips, x_ci=95)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# time 별로 분류
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='time')
```

# 사전정의된 그래프
    여러그래프들을 합쳐서 정보를 추출
    범주형 데이터에 적합

### 바이올린 플롯 : 상자그림과 분포도를 한번에 표시
    x : 범주형 데이터
    y : 연속형 데이터

```python
sns.violinplot(x='day', y='total_bill', hue='smoker',data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', hue='smoker',data=tips)
```

### FaceGrid: 영역을 나눠서 각각 다른 관점에서 범주형 데이터를 시각화
    전체 데이터를 범주형 데이터의 다양한 관점에서 나눠볼 수 있다

```python
g = sns.FacetGrid(data=tips, col='time', row='sex')
g.map(sns.scatterplot, 'total_bill','tip')
```
