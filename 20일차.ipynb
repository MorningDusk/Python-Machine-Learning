# 멀티 인덱스

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.rename(columns={'sex':'gender'}, inplace=True)
titanic.head()
```

```python
# 클래스별 성별 평균
classGender_grouped = titanic.groupby(['class', 'gender'])
```

```python
df = classGender_grouped.mean()
```

```python
# class가 first인 row들만 추출
df.loc['First']
```

```python
df.loc['First'].loc[['male']]
```

```python
df.loc[[('First', 'male')]]
```

```python
# gender 인덱스에서 male 값을 갖는 행을 추출
df.xs('male', level='gender')
```

# 피벗

```python
df = titanic.loc[:, ['age', 'gender', 'class', 'fare', 'survived']]
```

```python
df.head()
```

```python
df1 = pd.pivot_table(
    df,
    index = 'class',
    columns = 'gender',
    values = 'age',
    aggfunc = 'mean'
)
```

```python
df1
```

```python
df2 = pd.pivot_table(
    df,
    index = 'class',
    columns = 'gender',
    values = 'survived',
    aggfunc = ['mean', 'sum']
)
```

```python
df2
```

```python
# 피벗
# index class, gender
# column survived
# values age, fare
# aggfunc mean, max
df3 = pd.pivot_table(
    df,
    index = ['class', 'gender'],
    columns = 'survived',
    values = ['survived', 'fare'],
    aggfunc = ['mean', 'max']
)
df3
```

```python
df3.xs('First')
```

```python
df3.xs('female', level='gender')
```

```python
df3.xs(('First', 'female'))
```

```python
df3.xs('mean', axis=1)
```
