# 함수매핑
    Series객체.apply(매핑함수)

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
df = titanic[['age','fare']]
df['ten'] = 10
df.head()
# age fare
```

```python
def add_10(n):
    return n + 10
def add_two_obj(a,b):
    return a+b
```

```python
print(add_10(10))
print(add_two_obj(10,10))
```

```python
# apply df['age'] -> add_10 , add_two_obj
sr1 = df['age'].apply(add_10)
```

```python
sr2 = df['age'].apply(add_two_obj, b=10)
```

```python
sr1
```

```python
sr2
```

```python
# lambda를 이용해서 표현하기
sr3 = df['age'].apply(lambda x: add_10(x))
```

```python
sr3
```

    데이터 프레임 원소에 함수 매핑 - applymap()
    DataFrame 객체.apply(매핑함수)

```python
df = titanic.loc[:, ['age', 'fare']]
df_map = df.applymap(add_10)
df_map
```

    데이터 프레임 각 열에 함수 매핑
    DataFrame객체.apply(매핑함수, axis=0)
    데이터 프레임에 apply를 적용하면 모든 열을 하나씩 분리해서 매핑함수의 인자로 전달

```python
# 사용자 정의 함수
def missing_value(series):
    return series.isull()
```

```python
result = df.apply(add_10, axis=0)
result
```

```python
def min_max(x):
    return x.max() - x.min()
```

```python
df[['age']].apply(min_max, axis=0)
```

```python
df['ten'] = 10
df
```

```python
df['add'] = df.apply(lambda x : x['age'] + x['ten'], axis=1)
df.head()
```

    데이터 프레임객체에 함수 매핑 pip()
    데이터 프레임객체.pip(매핑함수)

```python
# 데이터 프레임을 전달하면 데이터프레임 반환
def missing_value(series):  #true, false
    return series.isnull()

# 데이터 프레임을 전달하면 시리즈를 반환
def missing_count(series):    #각 열의 NaN 개수
    return missing_value(series).sum()

# 데이터 프레임을 전달하면 값 반환
def total_number_missing(x):  #총 NaN의 개수
    return missing_count(x).sum()
```

```python
result_df = df.pipe(missing_value)
result_df.head()
```

```python
result_series = df.pipe(missing_count)
result_series
```

```python
df.pipe(total_number_missing)
```
