```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

plt.title('blobs data')
x, y = make_blobs(n_samples=300, n_features=2, centers=3, random_state=1)
plt.scatter(x[:,0], x[:,1], marker='o', c=y, s=100, edgecolors='k', linewidths=2)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

```python
from sklearn.datasets import make_moons

plt.title('moons data')
x,y = make_moons(n_samples=400, noise=0.1, random_state=0)
plt.scatter(x[:,0], x[:,1], marker='o', c=y, s=100, edgecolors='k', linewidths=2)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

```python
from sklearn.datasets import make_circles

plt.title('circles data')
x,y = make_circles(n_samples=500, shuffle=True, noise=0.05, random_state=None, factor=0.5)
plt.scatter(x[:,0], x[:,1], marker='o', c=y, s=100, edgecolors='k', linewidths=2)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```
