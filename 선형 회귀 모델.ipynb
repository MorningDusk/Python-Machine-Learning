```python
import mglearn
from sklearn.model_selection import train_test_split

mglearn.plots.plot_linear_regression_wave()
```

```python
from sklearn.linear_model import LinearRegression

x,y = mglearn.datasets.make_wave(n_samples=60)
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42)

print('Number of training data:', len(x_train))
lr = LinearRegression().fit(x_train,y_train)
```

```python
print(f"slope parameter: {lr.coef_}, intercept: {lr.intercept_}")
```

```python
print(f"학습 데이터셋 정확도: {lr.score(x_train, y_train)}, 테스트 데이터셋 정확도: {lr.score(x_test, y_test)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

line = np.linspace(-3, 3, 100).reshape(-1,1)

fig, ax = plt.subplots(figsize=(8,8))

ax.plot(line, lr.coef_ * line + lr.intercept_)

ax.spines['left'].set_position('center')
ax.spines['right'].set_color('gray')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_color('none')

ax.set_xlim(-3,3)
ax.set_ylim(-3,3)

ax.set_aspect('equal')

plt.grid()

plt.plot(x_train, y_train, 'o')
plt.ylim(-3,3)
plt.legend(['Linear model', 'Training data'], loc='best')
plt.show()
```
