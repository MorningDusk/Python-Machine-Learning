```python
import requests
import os

celestial_coordinates = []

f = open("C:/Users/user/python_machine_learning/2023_창의학술제/galclass_finding_list.txt", "r")
while True:
    data = f.readline()
    celestial_coordinates.append(data)
    if not data:
        break

new_coordinates = []

for coordinate in celestial_coordinates:
    new_coordinates.extend(coordinate.split())

ra_list = []
dec_list = []

for i in range(1, len(new_coordinates)-2, 3):
    ra_list.append(new_coordinates[(i)])
    dec_list.append(new_coordinates[(i+1)])

image_urls = []

for ra, dec in zip(ra_list, dec_list):
    image_urls.append("https://skyserver.sdss.org/dr16/en/tools/chart/navi.aspx?ra=" + ra + "&dec=" + dec + "&scale=0.4&width=120&height=120&opt=")
    

for i, image_url in enumerate(image_urls):
    try:
        image_data = requests.get(image_url).content
    except requests.exceptions.HTTPError as e:
        print(e)
    with open(os.path.join("F:/images/", f"{ra_list[i]}_{dec_list[i]}.png"), "wb") as f:
        f.write(image_data)
        print(f"{i + 1}/{len(image_urls)} 이미지 다운로드 완료")
```

```python
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import requests
import os
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/user/python_machine_learning/2023_창의학술제/GalaxyZoo1_DR_table_converted.csv")

ra_list = df['RA_decimal']
dec_list = df['DEC_decimal']

ra_list = list(map(str, ra_list))
dec_list = list(map(str, dec_list))

image_urls = []

spiral = df['SPIRAL']
elliptical = df['ELLIPTICAL']
uncertain = df['UNCERTAIN']

for ra, dec in zip(ra_list, dec_list):
    image_urls.append("https://skyserver.sdss.org/dr16/SkyServerWS/ImgCutout/getjpeg?TaskName=Skyserver.Chart.Navi&ra=" + ra + "&dec=" + dec + "&scale=0.049515875&width=1536&height=1536&opt=")

for i, image_url in enumerate(image_urls, start=77314):
    response = requests.get(image_url)
    if response.status_code == 200:
        image_data = response.content
        if spiral[i] == 1:
            with open(f"F:/images/spiral/{ra_list[i]}_{dec_list[i]}.jpg", "wb") as f:
                f.write(image_data)
            print(f"{i} / {len(image_urls)} spiral 이미지 다운로드 완료!")
        elif elliptical[i] == 1:
            with open(f"F:/images/elliptical/{ra_list[i]}_{dec_list[i]}.jpg", "wb") as f:
                f.write(image_data)
            print(f"{i} / {len(image_urls)} elliptical 이미지 다운로드 완료!")
        elif uncertain[i] == 1:
            with open(f"F:/images/uncertain/{ra_list[i]}_{dec_list[i]}.jpg", "wb") as f:
                f.write(image_data)
            print(f"{i} / {len(image_urls)} uncertain 이미지 다운로드 완료!")
    else:
        print(f"이미지 다운로드 실패. HTTP 상태 코드: {response.status_code}")
```
